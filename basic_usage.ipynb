While in .venv
##### Step 1: Register the venv as a Jupyter kernel
Run this while your venv is active:  


```bash
python -m ipykernel install --user \
    --name momp \
    --display-name "Python (.venv-momp)"
```

- name momp  internal kernel name
- display-name "Python (.venv-momp)"  the name you see in Jupyter
  
This will make your venv visible as a kernel.

##### Step 2: Check that Jupyter sees it  
```bash
jupyter kernelspec list
```

##### Step 3: Restart Jupyter
```
jupyter notebook
```
