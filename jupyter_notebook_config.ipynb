### Steps for setting up jupyter notebook for interactive MOMP applications:
#### For python virtual env:
##### Step 1: Install required packages
with .venv-momp active
```bash
pip install jupyterlab ipykernel
```
##### Step 2: Register the venv as a Jupyter kernel
```bash
python -m ipykernel install --user \
    --name venv-momp \
    --display-name "Python (.venv-momp)"
```
- name momp → internal kernel name
- display-name "Python (.venv-momp)" → the name you see in Jupyter
  
This will make your venv visible as a kernel.

##### Step 3: Check that Jupyter sees it  
```jupyter kernelspec list```

you should see
```
python3
venv-momp
```
##### Step 4: Launch Jupyter
```
jupyter notebook
```
then select
`Kernel → Change Kernel → Python (.venv-momp)`
##### Step 5: Verify inside the notebook
```
import sys
print(sys.executable)
```
you get the full path to the Python interpreter that is running the current Jupyter kernel
`/home/user/miniconda3/envs/myenv/bin/python`
or on **Windows**
`C:\Users\You\miniconda3\envs\myenv\python.exe`
